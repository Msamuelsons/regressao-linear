```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

```python
#gerando uma massa de dados
x, y = make_regression(n_samples=300, n_features=1, noise=30)

#mostrando o gr√°fico
plt.scatter(x, y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

#cria√ß√£o do modelo
modelo = LinearRegression()
```

```python
modelo.fit(x, y)
```

```python
coef_linear = modelo.intercept_ #coeficiente linear
```

```python
#y = mx + b
#m = 91.45 ou modelo.intercept_
#b = modelo.coef
coef_angular = modelo.coef_#coeficiente angular
```

```python
#mostrando resultado
import numpy as np
plt.scatter(x, y)

xreg = np.arange(-3, 3, 1)
plt.plot(xreg, coef_angular*xreg+coef_linear, color='black') #gr√°fico regress√£o
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.30) # pega 30% do total de dados para teste
```

```python
modelo.fit(x_treino, y_treino)
```

```python
resultado = modelo.score(x_teste, y_teste) # coeficiente de determina√ß√£o
```

```python
print(resultado * 100)
```

```python
plt.scatter(x_teste, y_teste)

xreg = np.arange(-3, 3, 1)
plt.plot(xreg, coef_angular*xreg+coef_linear, color='black') #gr√°fico regress√£o
plt.show()
```
